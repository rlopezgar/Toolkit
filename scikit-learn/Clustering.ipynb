# Agrupación
La agrupación o clustering es un conjunto de métodos de aprendizaje no supervisado que busca hallar similitudes en los elementos de un conjunto de datos sin etiquetar para dividirlos en grupos o clusters. 
  

Algunos tipos de agrupación son las siguientes:
* Agrupación basada en centroides  
Se agrupan los puntos con base en su distancia a puntos dados denominados centroides que no necesariamente están en el conjunto de datos en cuestión.  
* Agrupación jerárquica  
Se agrupan los puntos basados en su proximidad. A diferentes distancias se harán diferentes grupos que pueden ser expresados jerárquicamente mediante un dendograma. Se trata de contruir una jerarquía de clusters donde cada nodo es un clúster del cuál se desprenden otros clústers. Hay dos tipos de agrupación jerárquica. Por un lado, está el enfoque **divisivo** en que la partición de los clusters es descendente por lo que se inicia con todas las observaciones en un cluster grande y se dividen hacia clusters más pequeños dependiendo de los valores que toman las variables de los elementos en cuestión. Por otro lado está el enfoque **aglomerativo** en el que la partición de los clusters es ascendente por lo que cada observación es un cluster pequeño y se aglomeran hacia clusters más grandes dependiendo de los valores que toman las variables de los elementos en cuestión.  
* Agrupación basada en densidad  
Se agrupan los puntos por áreas con mayor densidad. Los puntos que están en áreas dispersas se suelen considerar como ruido o outliers.  
* Agrupación basada en distribución  
Se agrupan los puntos con base en qué tan probable es que pertenezcan a cierta distribución.

## K-Means
K-medias es un algoritmo de aprendizaje no supervisado del tipo agrupación basada en centroides. Los objetos dentro de una partición son muy similares y los objetos de las distintas particiones son muy diferentes o disimilares. El algoritmo funciona tratando de minimizar la distancia de los elementos dentro del mismo cluster y maximizar la distancia de los elementos de los cluster diferentes empezando por tomar $k$ centros aleatorios de clusters denominados centroides. Note que, dado que el algortimo trabaja con distancias ente puntos, es menester normalizar los datos. 

El algortimo tiene el siguiente funcionamiento.
Considere $x_1,...,x_n \in\mathbb{R}^m$.
1. Elija el número $k$ de clusters con sus respectivos centroides $c_1, c_2, ..., c_k$ dentro o fuera de la muestra.
2. Obtenga la distancias de cada $d(x_i,c_1), d(x_i,x_2),...,d(x_i,c_k)$ $\forall 1\leq i \leq n$ en orden de tener una matriz de distancias $M\in\mathcal{M}_{n\times k}(\mathbb{R})$.
3. Asigne a $x_i$ el centroide más cercano.
4. Para cada cluster, asigne un nuevo centroide dado por la media de los puntos que los componen, coordenada a coordenada. Esto es, suponga que para el $j$-ésimo centroide $c_j$ $(1\leq j \leq k)$ se le tienen asignados $a$ puntos $x_{j,1},...,x_{j,a}$ donde $x_{j,l} = (x_{j,l,1},...,x_{j,l,m})$ con $1\leq l \leq a$ así pues, el nuevo $j$-ésimo centroide será $c_j = (\sum_{l=1}^a x_{j,l,1}, ...,\sum_{l=1}^a x_{j,l,m})$.  
6. Repita 2,3,4 hasta que los centroides no cambien.


    
Para implementar el algortimo KNN con las columnas col_1,...,col_n de una tabla df dado el hiperparámetro $k$,  utilice el siguiente código.  
1. Convertir los datos a un arreglo de NumPy.  
`data = df.to_numpy(dtype = float)`  
2. Estandarizar los datos usando Z-score o algún otro normalizador.  
`from sklearn.preprocessing import StandardScaler`  
`scaler = StandardScaler()`  
`data = scaler.fit_transform(data)`
3. Entrenar al modelo.  
Tome en cuanta los sigueintes parámetros.
* `init` es el método de algoritmo k-means a usar. Se usará `init = 'k_means++'` puesto que éste selecciona centroides de clusters iniciales usnado muestreo basado en la distribución de probilidad empirica de los puntos y su contribución a la inercia, lo que acelera la convergencia. Sin embargo, para iniciarlos aleatoriamente puede usar `init = rand`.  
* `clusters` es el número de clusters.  
* `n_init` es el número de veces que el algoritmo se corre con diferentes semillas aleatorias. Esto pues si bien el algoritmo converge, puede llegar a un óptimo local y no a uno global. Puede ser necesario inicializar aleatoriamente los centroides en diferentes puntos y repetir el proceso para ver que se llegue al mejor resultado. Se dejará en `n_init = 10`, el default para `init = rand`.

`from sklearn.cluster import KMeans`  
`# Crear un onjeto KNN`  
`km_model = KMeans(init = 'k-means++', n_clusters = k, n_init = 10)`  
`# Entrenar al modelo`  
`km_model.fit(data)`    
4. Hacer una predicción.  
`km_model.predict(data)`
5. Obtener los centroides.    
`km_model.cluster_centers_`

`labels = km_model.labels_`  
`df['labels'] = labels`  
7. Analizar las variables agrupando por cada cluster para nombrarlos.
`df.groupby('labels').agg(['count','min', 'max','mean'])`

**Método del codo**  
Se calcula la distancia de los puntos de cada cluster a su repectivo centroide, se eleva al cuadrado y se suma. A esto se le denomina inercia. Esto es, considere los $k$ centroides $c_1,...,c_k$  cuyo $j$-ésimo elemento $c_j$ $(1\leq j \leq k)$ tiene a los $a$ puntos $x_{j,1},...,x_{j,a}$. Así pues, la incercia está dada como sigue:
$$Inercia = \sum_{j = 1}^k\sum_{l=1}^a d(x_{j_l}, c_j)^2$$
Si hay  más centroides, entonces habrá menos inercia; si nembarog,  hay un punto denominado 'codo' en el que después de éste la incercia cae dramáticamente. El valor de $k$ del codo suele ser la mejor para realizar la agrupación. La elección se toma mediante la visualiuzación de la 'gráfica de codo' donde en el eje $x$ representa a los diferentes valores de $k$ y el eje $y$ representa la inercia. Use el siguiente código: 

`def km_elbow_plot(data):`  
`    sse = {}`  
`    for k in range(1, 10):`  
`        km_model = KMeans(init = "k-means++", n_clusters = k, n_init = 10)`   
`        km_model.fit(data)`    
`        sse[k] = km_model.inertia_`  
`    plt.plot(list(sse.keys()), list(sse.values()))`  
`    plt.xlabel('$k$')`  
`    plt.ylabel('Inertia')`  
`km_elbow_plot(data)`
  
**Método de la silueta**  
Se define el Silhouette Score $SH$ como sigue:
$$SC = \frac{b-a}{max(a,b)}$$ 
donde:  
* $b$ es el promedio de las distancias entre los elementos de la muestra y el cluster más cercano al cual no pertenezcan.
* $a$ es el promedio de las distancias entre los elementos de la muestra y el cluster al cual pertencen.

Note que $-1\leq SH \leq 1$ y se interpreta de la sigueinte manera:
* Los mejores valores son cercanos a 1.
* Los peores valores son cercanos a -1. De hecho, valores negativos generalmente indican que un elemento de la muestra ha sido asignado al cluster erróneo puesto que un cluster se ajusta más a éste.
* Valores cercanos a cero indican que hay clusters que se traslapan.
Así pues, el método de la silueta consiste en tomar el hiperparámetro $k$ tal que se maximice $SH$. La elección se toma mediante la visualiuzación de la 'gráfica de silueta' donde en el eje $x$ representa a los diferentes valores de $k$ y el eje $y$ representa los valores de $SH$. Use el siguiente código: 

`from sklearn.metrics import silhouette_score`  
`def km_silhouette_plot(data):`  
`    slhtte = {}`  
`    for k in range(2, 10):`  
`        km_model = KMeans(init = "k-means++", n_clusters = k, n_init = 10)`   
`        km_model.fit(data)`    
`        slhtte[k] = silhouette_score(data, km_model.labels_)`
`    plt.plot(list(sse.keys()), list(sse.values()))`  
`    plt.xlabel('$k$')`  
`    plt.ylabel('Silhouette Score')`  
`km_silhouette_plot(data)`




Notas:
* Se requiere definir una cantidad preestablecida de clusters, pero esto es optimizable tunneando éste hiperparámetro. 
* Úsese con datos numéricos en los que tenga sentido considerar la distancia.
* Los resultados pueden cambiar dependiendo la noción de distancia que se utilice en el parámetro `metric` que por default toma distancia euclidiana.

```python
# Elegir la mejor k por método del codo
def km_elbow_plot(data):
    sse = {}
    for k in range(2, 10):
        km_model = KMeans(init = "k-means++", n_clusters = k, n_init = 10)   
        km_model.fit(data)
        sse[k] = km_model.inertia_
    plt.plot(list(sse.keys()), list(sse.values()))
    plt.xlabel('$k$')
    plt.ylabel('Inertia')
km_elbow_plot(data)
# Use k = 4
```

```python
# Elegir la mejor k por método de la silueta
from sklearn.metrics import silhouette_score
def km_silhouette_plot(data):
    slhtte = {}
    for k in range(2, 10):
        km_model = KMeans(init = "k-means++", n_clusters = k, n_init = 10)   
        km_model.fit(data)
        slhtte[k] = silhouette_score(data, km_model.labels_)
    plt.plot(list(slhtte.keys()), list(slhtte.values()))
    plt.xlabel('$k$')
    plt.ylabel('Silhouette Score')
km_silhouette_plot(data)
# Use k = 4
```

## Cúmulos
Cúmulos es un algoritmo de aprendizaje no supervisado del tipo agrupación jerárquica con enfoque aglomerativo. Suponga que tiene una muestra de $k$ elementos con $l$ características que se buscan agrupar de acuerdo a su proximidad en clusters. Note que, dado que el algortimo trabaja con distancias ente puntos, es menester normalizar los datos. 
  
Sean $F$, $G$ dos grupos de individuos. Note que en un principio habrían $k$ clusters conformados por los individos mismos. Suponga que $|F|=n$ y $|G|=m$. Se procede al cálculo de distancias de acuerdo a alguno de los siguientes criterios:  
* Ditancia mínima (Single-Linkage Clustering)
$$d(F,G) = min\{d(x,y):x\in F, y\in G\}$$
* Distancia máxima (Complete-Linkage Clustering)
$$d(F,G) = max\{d(x,y):x\in F, y\in G\}$$
* Distancia promedio (Average Linkage Clustering)
$$d(F,G) = \sum_{i=1}^n\sum_{j=1}^m \frac{d(x_i,y_j)}{nm}$$
  
donde para $x_i = (x_{i,1},...,x_{i,l})$ y $x_j=(x_{j,1},...,x_{j,l})$ $(1\leq i,j \leq k)$ se tiene que:
$$d(x_i,x_j) = \sqrt{\sum_{a = 1}^l (x_{i,a} - x_{j,a})^2}$$

Así pues, suponga que se tienen $k$ individuos cuyas distancias están dadas en una matriz simétrica $M_k(\mathbb{R})$ tal que $diag(M) = \bar{0}$. 

El algoritmo tiene el siguiente funcionamiento:
1. Considere $M$.
2. Agrupe los individuos con menores distancias entre sí.
3. Obtenga una nueva matriz de distancias considerando los nuevos grupos y note que tendrá dimensión estrictamente menor a $k$.
4. Repita 1,2,3 hasta llegar a formar un solo grupo con una matriz de distancia simétrica de diagonal $\bar{0}$ y dimensión 1. 
  
Para implementar el algortimo de cúmulos para agrupar en $k$ clusters con las columnas col_1,...,col_n de una tabla df utilice el siguiente código.  
1. Convertir los datos a un arreglo de NumPy.  
`data = df.to_numpy(dtype = float)`

`from sklearn.preprocessing import StandardScaler`  
`scaler = StandardScaler()`  
`data = scaler.fit_transform(data)`  
3. Entrenar al modelo.
Considere los siguientes parámetros:
* `n_clusters` se trata de la cantidad de clusters a definir.
* `linkage` se trata de el criterio de distancia a tomar. El desault es `linkage = 'ward'`y  minimiza la varianza de los clusters que se unen. `linkage = ‘average'` toma la distancia promedio, `linkage = ‘complete’` toma la distancia máxima y `linkage = ‘single’` toma la distancia mínima.  
`# Importar las herramientas requeridas`  
`from sklearn.cluster import AgglomerativeClustering`  
`# Crear un objeto AgglomerativeClustering`  
`agglom_model = AgglomerativeClustering(n_clusters = 4, linkage = 'average')`  
`# Entrenar al modelo`  
`agglom_model = agglom_model.fit(data)`  
4. Obtener las etiquetas de cluster asignadas respectivamente a cada punto y ponerlas en la tabla df.  
`labels = agglom_model.labels_`  
`df['labels'] = labels`  
5. Analizar las variables agrupando por cada cluster para nombrarlos.
`df.groupby('labels').agg(['count', 'min', 'max','mean'])`
6. Dibujar el dendograma.  
`import scipy as sp`   
`dist_matrix = sp.spatial.distance_matrix(data, data)`  
`link_matrix = sp.cluster.hierarchy.linkage(dist_matrix, 'complete')`  
`dendro = sp.cluster.hierarchy.dendrogram(link_matrix)`  

Notas:
* Se requiere definir una cantidad preestablecida de clusters.
* Úsese con datos numéricos en los que tenga sentido considerar la distancia. 
*  Los resultados pueden cambiar dependiendo la noción de distancia que se utilice en el parámetro `metric` que por default toma distancia euclidiana.
*  Los resultados pueden cambiar dependiendo el criterio de distancia entre clustersque se utilice en el parámetro `linkage` que por default toma minimización de varianza.
* Úsese en conjuntos de datos relativamente pequeños, en orden de que el dendograma no resulte confuso y considerando que el algortimo tiene un alto costo computacional.
* Para predecir hay que reentrenar.
* No hay una manera de evaluar desempeño.

## DBSCAN 
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) es un algoritmo de aprendizaje no supervisado del tipo basado en densidad. Suele ser utilziado para datos espaciales o con formas arbitrarias. El algortimo localiza regionones de alta densidad que están separadas unas de otras por regiones de baja densidad, donde se entiende densidad como el número de puntos en un radio especificado; más aún, de esta forma se ignoran el ruido o los puntos anómalos. La noción fundamental del algoritmo es que si un elemento de la muestra dada pertenece a un cluster es porque debería estar cerca de muchos otros puntos que perteneces a ese cluster. Note que, dado que el algortimo trabaja con distancias ente puntos, es menester normalizar los datos. 
  
Para definir un cluster el DBSCAN utiliza dos parámetros $r$ y $m$. Una vecindad de puntos se considera un cluster si tiene al menos un radio $r$ en el que se incluyen al menos $m$ puntos. Se clasificarán a los puntos como sigue:
* Un punto es core si dentro de su $r$-vecindad hay al menos $m$ puntos.
* Un punto es border si dentro de su $r$ vecindad no hay al menos $m$ puntos o si su distancia a un punto core no es mayor a $r$.
* Un punto es outlier si no es un punto central ni está lo suficientemente cerca de uno como para ser un punto border.
Unavez etiquetados los puntos, estos se agruparán en sus resepctivos clusters.

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score 
from itertools import product
def dbscan_silhoutte_hm(data, min_m, max_m):
    r_vals = np.arange(0.1,1,.1)
    m_vals = np.arange(min_m, max_m)
    dbscan_params = list(product(r_vals, m_vals))
    num_clusters = []
    sil_score = []
    for p in dbscan_params:
        dbscan_model = DBSCAN(eps=p[0], min_samples=p[1]).fit(data)
        num_clusters.append(len(np.unique(dbscan_model.labels_)))
        sil_score.append(silhouette_score(data, dbscan_model.labels_))
    tmp_cluster = pd.DataFrame.from_records(dbscan_params, columns = ['r', 'm'])
    tmp_cluster['num_clusters'] = num_clusters
    pivot_cluster = pd.pivot_table(tmp_cluster, values = 'num_clusters', index = 'm', columns = 'r')
    tmp_sil = pd.DataFrame.from_records(dbscan_params, columns = ['r', 'm'])
    tmp_sil['sil_score'] = sil_score
    pivot_sil = pd.pivot_table(tmp_sil, values = 'sil_score', index = 'm', columns = 'r')
    return pivot_cluster, pivot_sil
p_c, p_s = dbscan_silhoutte_hm(data, 5,15)
```
