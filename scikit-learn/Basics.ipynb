### Normalización
Se trata de llevar los datos a un formato tal que sea más intuitivo su análisis o más preciso su procesamiento.

#### Simple feature scaling  
Esto hace que los nuevos valores oscilen entre 0 y 1.  
$x_{new} = \frac{x_{old}}{x_{max}}$  
Para aplicar simple feature scaling a una columna col de una tabla df:  
`df['col_i'] = df['col']/df['col'].max()`  
Ahora bien, si desea aplicar una estandarización inversa a una columna col de una tabla df:  
`df['col_i'] = df['col']/df['col'].min()`  

#### Min-Max  
Esto hace que los nuevos valores oscilen entre 0 y 1.  
$x_{new} = \frac{x_{old} - x_{min}}{x_{max} - x_{min}}$  
Para aplicar min-max a una columna col de una tabla df:  
`df['col'] = (df['col'] - df['col'].min())/(df['col'].max() - df['col'].min())`  
O bien, usando sklearn,  
`from sklearn.preprocessing import MinMaxScaler`  
`scaler = MinMaxScaler()`  
`df['col'] = scaler.fit_transform(df[['col']])`  
Más aún, para estandarizar todas las columas de la tabla df:  
`from sklearn.preprocessing import MinMaxScaler`  
`scaler = MinMaxScaler()`   
`df = scaler.fit_transform(df)`  
Ahora bien, si desea aplicar una estandarización inversa a una columna col de una tabla df dado que ya fue estandarizada mediante un objeto StandardScaler() scaler:  
`df['col'] = scaler.inverse_transform(df[['col']])`  
Más aún, para desestandarizar todas las columnas de la tabla df dado que ya fue estandarizada mediante un objeto StandardScaler() scaler:  
`df = scaler.inverse_transform(df)`  

#### Z-score  
$\mu$ es la media del conjunto de datos y $\sigma$ es la desviación estándar del mismo. Esto haceque los nuevos valores ronden el 0 y usualmente oscilan entre tres positivos y tres negativos, aunque podrían ser más bajos o más altos.  
$x_{new} = \frac{x_{old} - \mu}{\sigma}$  
Para aplicar z-score a una columna col de una tabla df:  
`df['col'] = (df['col'] - df['col'].mean())/df['col'].std()`  
O bien, usando sklearn,  
`from sklearn.preprocessing import StandardScaler`  
`scaler = StandardScaler()`   
`df['col'] = scaler.fit_transform(df[['col']])`  
Más aún, para estandarizar todas las columas de la tabla df:  
`from sklearn.preprocessing import StandardScaler`  
`scaler = StandardScaler()`   
`df = scaler.fit_transform(df)`  
Ahora bien, si desea aplicar una estandarización inversa a una columna col de una tabla df dado que ya fue estandarizada mediante un objeto StandardScaler() scaler:  
`df['col'] = scaler.inverse_transform(df[['col']])`  
Más aún, para desestandarizar todas las columnas de la tabla df dado que ya fue estandarizada mediante un objeto StandardScaler() scaler:  
`df = scaler.inverse_transform(df)`  

#### Binning  
Se trata de agrupar los valores en contenedores.  
Para agrupar una columna col en n grupos g_1, g_2, ... g_n con etiquetas et_1,...,et_n:  
`bins = np.linspace(min(df['col']), max(df['col']), n+1)`  
`group_names = ['et_1', 'et_2', ..., 'et_n']`  
`df['col_binned'] = pd.cut(df['col'], bins, labels = group_names, include_lowest = True)`

#### One-Hot Encoding
Se trata de mapear variables categóricas a  valores numéricos en orden de que estos puedan ser procesados por el modelo. Para esto se usa la técnica dneominada one-hot encoding. Por cada categoría cat_1, cat_2, ..., cat_n se hará una nueva columna con True en caso de que el renglón en cuestión posea esa característica y con False en caso de que el renglón en cuestión no posea esa caracteríostica.  
  
Para aplicar one-hot encoding a una columna col con valores categóricos:   
`# Ver cuáles son las categorías de la columna`  
`print(df['col'].unique())`  
`# Generar una tabla dummies con el one-hot encoding`  
`dummies = pd.get_dummies(df['col'])`  
`# Renombrar las columas se la tabla generada`  
`dummies = dummies.rename(columns={'cat_1':'cat_1_name', 'cat_2':'cat_2_name',..., 'cat_n': 'cat_n_name')`  
`# Unir la tabla original con la tabla generada`  
`df = pd.concat([df, dummies], axis=1)`  
`# Eliminar la columna col`  
`df = df.drop('col', axis = 1)`  

### Enumerado de etiquetas
Se trata de sustituir etiquetas en string por etiquetas en int para una variable. Esto es, para una tabla df con una variable col que toma valores et_1, et_2,...,et_n se sustituye respectivamente  por 1,2,...,n. 
  
Para aplicarle enumerado de etiquetas a col use el siguiente código:  
`from sklearn import preprocessing`  
`col_ee = preprocessing.LabelEncoder()`  
`col_ee.fit(['et_1','et_2',...,'et_n'])`  
`df['col'] = col_ee.transform(df['col'])`

#### Correlación de Pearson
Se trata de analizar relaciones lineales entre variables numéricas y la variable objetivo (numérica). Para esto utilizaremos dos datos numéricos:
* Coeficiente de correlación $\varphi$ donde $\varphi_{X,Y} = \frac{cov(X,Y)}{\sigma_X\sigma_Y} = \frac{(\sum_{i=1}^nx_i-\bar{x})(y_i-\bar{y})}{(\sqrt{\sum_{i=1}^n(x_i-\bar{x})^2})(\sqrt{\sum_{i=1}^n(y_i-\bar{y})^2})}$
* p-value $p$
Los coeficientes de correlación serán interpretados de la siguiente manera:
* Si $\varphi \approx 1$, entonces se observa una relación lineal postiva
* Si $\varphi \approx 1$, entonces se observa una relación lineal negativa.
* Si $\varphi \approx 0$, entonces no se observa una relación lineal.
A su vez, esto será corroborado de la siguiente manera:
* Si $p \leq 0.001$, entonces hay certeza fuerte del resultado. 
* Si $p \leq 0.05$, entonces hay certeza moderada del resultado.
* Si $p \leq 0.01$, entonces hay certeza débil del resultado.
* Si $p > 0.01$, entonces no hay certeza del resultado.
  
Diremos que hay una correlación significativa cuando $\varphi \approx 1$ o $\varphi \approx -1$ y $p \leq 0.001$, $p \leq 0.05$ o $p \leq 0.01$.  

Para obtener el coeficiente de  correlación entre dos columnas col y la columna de variable objetivo col_obj con su respectivo p-value:  
`import scipy as sc`  
`corr_coef, p_value = sc.stats.pearsonr(df['col'], df['col_obj'])`

#### Análisis de varianza (ANOVA)
El Análisis de varianza (ANOVA) es un método estadístico utilizado para evaluar si existen diferencias significativas entre las medias de dos o más grupos.  
  
Para esto utilizaremos dos datos numéricos:  
* Puntuación del F-test $F$. 
* p-value $p$  
La puntuación F serán interpretados de la siguiente manera:
* Una F grande indica una relacion lineal fuerte entre las categorías de la variable y la variable objetivo.
* Una F chica indica una relacion lineal débil entre las categorías de la variable y la variable objetivo.
  
A su vez, esto será corroborado de la siguiente manera:
* Si $p \leq 0.001$, entonces hay certeza fuerte del resultado. 
* Si $p \leq 0.05$, entonces hay certeza moderada del resultado.
* Si $p \leq 0.01$, entonces hay certeza débil del resultado.
* Si $p > 0.01$, entonces no hay certeza del resultado.
  
Diremos que hay una correlación lineal significativa cuando $F$ sea lo suficientemente grande y $p \leq 0.001$, $p \leq 0.05$ o $p \leq 0.01$. 
  
Para obtener la puntuación $F$ en una columna col con categorías cat_1,...,cat_n y la variable objetivo col_obj con su respectivo p-value:  
`import scipy as sc`  
`df_anova = df[['col', 'col_obj']]`  
`grouped_anova = df_anova.groupby(['col'])`  
`f_val, p_val = sc.stats.f_oneway(grouped_anova.get_group('cat_1')['col_obj'],..., grouped_anova.get_group('cat_n')['col_obj'])`

## Definición de conjunto de entrenamiento y conjunto de evaluación
Se hace una vez que se ha hecho el EDA. A sabiendas de cuales son las variables dependientes que podrían ser buenos predictores de la variable objetivo se dividirá el conjunto de datos como sigue:
* Conjunto de entrenamiento o training set
* Conjunto de evaluación o test set

Usamos el **conjunto de entrenamiento** o training set para crear un modelo y descubrir relaciones de predicción. Después utilizamos un **conjunto de evaluación** o test set para evaluar el comportamiento del modelo. **Cuando hayamos terminado de probar nuestro modelo, debemos usar todos los datos para entrenar el modelo**. 
  
Para dividir aleatoriamente un conjunto de datos df con variables predictoras col_1, col_2, ..., col_n y variable a predecir col_obj en conjuntos de entrenamiento x_train, y_train y conjuntos de prueba x_test, y_test en proporción $p$ para el entrenmiento y $1-p$ para prueba $(0 < p < 1 )$ utilice el siguiente código:  
`from sklearn.model_selection import train_test_split`  
`x_train, x_test, y_train, y_test = train_test_split(df[['col_1','col_2',...,'col_n']], df['col_obj'], test_size = 1-p)`

### sklearn.datasets.make_blobs
Se trata de una herramienta de scikit-learn que nos permite generar conjutnos de datos bidimensionales. Se trata de $n$ datos agrupados en $k$ clusters con centroides en $(x_1,y_1), (x_2,y_2),...,(x_k, y_k)$ y desviación estándar para cada cluster std en un arreglo x con su respetiva etiqueta en un arreglo y. Para realziarlo, utilice el siguiente código:
  
`import sklearn`  
`from sklearn.datasets import make_blobs`  
`x,y = make_blobs(n_samples = n, centers = [[x_1,y_1], [x_2,y_2],...,[x_k, y_k]], cluster_std = std)`

```python
import pandas as pd 
import sklearn
import seaborn as sns
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples = 5000, centers=[[4,4], [-2, -1], [2, -3], [1, 1]], cluster_std = 0.9)
df_xy = pd.DataFrame(x, columns = ['x_axis', 'y_axis'])
df_et = pd.DataFrame(y, columns = ['et'])
df = pd.concat([df_xy, df_et], axis = 1)
sns.scatterplot(data = df, x = 'x_axis', y = 'y_axis', hue = 'et')
```

## Persistencia de modelo
Si se cierra una sesión sin haber guardado un modelo ya entrenado, entonces al abrir una nueva sesión se tendrá que reentrenar. Así pues, para su reutilización posterior o persistencia se requiere guardarlo. 

Para guardar un modelo entrenado ml_model con el nombre ml_model_name.joblib use el siguiente código:  
`from joblib import dump, load`  
`dump(ml_model, 'ml_model_name.joblib')`    
Para abrir el modelo entrenado guardado como ml_model_name.joblib y asignarlo a un objeto de nombre ml_model en una sesión nueva use el siguiente código:   
`from joblib import dump, load`  
`ml_model = load('ml_model_name.joblib')`
