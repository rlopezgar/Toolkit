## Logistic Regression
When working with PyTorch we'll always estimate the parameters of a model by a minimizing the cost of a loss function. For logistic regression or binary classification models, binary cross entropy loss (nn.BSELoss) is tipically used. For softmax regression or multi-class classification models, cross entropy loss (nn.CrossEntropyLoss) is tipically used. 
  
There are two ways in order to create a logistic regression model for certain input_size use the following code:  
  
**nn**  
`import torch.nn as nn`  
`class logistic_regression(nn.Module):`  
` `  
`    def __init__(self, input_size):`  
`        super(logistic_regression, self).__init__()`  
`        self.linear = nn.Linear(input_size, 1)`  
` `    
`    def forward(self, x):`  
`        yhat = torch.sigmoid(self.linear(x))`  
`        return yhat`  
` `  
`model = logistic_regression(n_inputs)`  

In order to train a logistic regression model, no matter if it's from nn.module or a custom one, follow the next steps:
    
0. Import required tools
1. Create a Dataset object.
2. Create model object.
3. Crete a objects for training (training function, criterion object, DataLoader object, optimizer).
4. Train model.
  
Then, in order to get the classes corresponding to each sample with it's respective probabily do as follows:
  
`# Get probabilities  ( of belonging to class 1 )`  
`probs = model(data_set.x)`  
`probs = probs.detach().numpy().round(5)`

`probs`

`labels = probs > 0.5`

```python
## 0. Import required tools
from torch.utils.data import Dataset, DataLoader
from torch import nn, optim
```

```python
## 1. Create Dataset object

# Dataset class
class Data(Dataset):
    
    # Constructor
    def __init__(self):
        self.x = torch.arange(-1, 1, 0.1).view(-1, 1)
        self.y = torch.zeros(self.x.shape[0], 1)
        self.y[self.x[:, 0] > 0.2] = 1
        self.len = self.x.shape[0]
        
    # Getter
    def __getitem__(self, index):      
        return self.x[index], self.y[index]
    
    # Get Length
    def __len__(self):
        return self.len

# Dataset object
data_set = Data()

# Plot data
X = data_set[:][0]
Y = data_set[:][1]
plt.plot(X[Y == 0].numpy(), Y[Y == 0].numpy(), 'bo', label = 'y = 0')
plt.plot(X[Y == 1].numpy(), 0 * Y[Y == 1].numpy(), 'ro', label = 'y = 1')
plt.ylim((-0.1, 3))
plt.legend()
```

```python
## 2. Create a model object

# Model class
class logistic_regression(nn.Module):
    
    def __init__(self, n_inputs):
        super(logistic_regression, self).__init__()
        self.linear = nn.Linear(n_inputs, 1)
        
    def forward(self, x):
        yhat = torch.sigmoid(self.linear(x))
        return yhat

# Model object
model = logistic_regression(1)

# View initial parameters
print("The parameters: ", list(model.parameters()))
```

```python
## 3. Create objects for training

# Trainer function
def train_model(criterion, train_loader, optimizer, epochs):    
    LOSS = []
    loss_epoch = 0
    for epoch in range(epochs):
        for x,y in train_loader:
            yhat = model(x)
            loss = criterion(yhat, y)
            loss_epoch = loss_epoch + loss
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        LOSS.append(loss_epoch.item())
        loss_epoch = 0
    return LOSS 

# Criterion object 
criterion = nn.BCELoss()

# DataLoader object 
train_loader = DataLoader(dataset = data_set, batch_size = 5)

# Optimizer
optimizer = optim.SGD(model.parameters(), lr = 0.1)
```

```python
## 4. Train model
LOSS = train_model(criterion, train_loader, optimizer, epochs = 100)

# View Loss
plt.plot(LOSS)
plt.xlabel("Epochs")
plt.ylabel("Cost")
```

```python
## Get probabilities (of belonging to class 1)

# Get probabilities
probs = model(data_set.x)
# Trun tensor into a numpy array use detach to not require grad
probs = probs.detach().numpy().round(7)
# Print in a readable way
np.set_printoptions(suppress = True)
probs
```

```python
## Get labels (True for class 1 and False for class 0)
labels = probs > 0.5
labels
```

```python
# Get accuracy
probs = model(data_set.x)
label = probs > 0.5
print("The accuracy: ", torch.mean((label == data_set.y.type(torch.ByteTensor)).type(torch.float)))
```

## Softmax Regression
When working with PyTorch we'll always estimate the parameters of a model by a minimizing the cost of a loss function. For softmax regression or multi-class classification models, cross entropy loss (nn.CrossEntropyLoss) is tipically used. 
  
There are two ways in order to create a softmax regression model for certain input_size and output_size, the following code: 
  
**nn**  
`import torch.nn as nn`  
`class SoftMax(nn.Module):`  
` `  
`    def__init__(self, in_size, out_size):`  
`    super(Softmax, self).__innit__()`  
`    self.linear = nn.Linear(in_size, out_size)`  
` `  
`    def forward(self, x):`  
`        out = self.linear(x)`  
`        return out`  
` `  
`model = SoftMax(input_size, output_size)`  

In order to train a logistic regression model, no matter if it's from nn.module or a custom one, follow the next steps:
    
0. Import required tools
1. Create a Dataset object.
2. Create model object.
3. Crete a objects for training (training function, criterion object, DataLoader object, optimizer).
4. Train model.
  
Then, in order to get the classes corresponding to each sample with it's respective probabily do as follows:
  
`# Get labels`  
`z =  model(data_set.x)`  
`_, labels = z.max(1)`  
`labels`  
`# Get probabilities of belonging to each class`  
`f_softmax = nn.Softmax(dim = -1)`  
`probs = f_softmax(z)`  
`probs = probs.detach().numpy().round(5)`  
`np.set_printoptions(suppress = True)`  
`probs`  

Notice that instead of 1, we'll use a output_size parameter in the constructor of this class. The parameter corresponds to the number of classes in the output. In addition there is no logistic funtion on the output. We have the column indexes side by side to the actual values of z on the right. We call the max function on z, this will give us the label. Then, applying the softmax function, we'll get the propability of belonging to each class.

### Softmax Regression for numerical data classification

```python
## 0. Import required tools
from torch.utils.data import Dataset, DataLoader
from torch import nn,optim
```

```python
## 1. Create a Dataset object

# Dataset class
class Data(Dataset):
    
    # Constructor
    def __init__(self):
        self.x = torch.arange(-2, 2, 0.1).view(-1, 1)
        self.y = torch.zeros(self.x.shape[0])
        self.y[(self.x > -1.0)[:, 0] * (self.x < 1.0)[:, 0]] = 1
        self.y[(self.x >= 1.0)[:, 0]] = 2
        self.y = self.y.type(torch.LongTensor)
        self.len = self.x.shape[0]
        
    # Getter
    def __getitem__(self,index):      
        return self.x[index], self.y[index]
    
    # Get Length
    def __len__(self):
        return self.len

# Dataset object
data_set = Data()

# View data
X = data_set[:][0]
Y = data_set[:][1]
plt.plot(X[Y == 0, 0].numpy(), Y[Y == 0].numpy(), 'bo', label = 'y = 0')
plt.plot(X[Y == 1, 0].numpy(), 0 * Y[Y == 1].numpy(), 'ro', label = 'y = 1')
plt.plot(X[Y == 2, 0].numpy(), 0 * Y[Y == 2].numpy(), 'go', label = 'y = 2')
plt.ylim((-0.1, 3))
plt.legend()
```

```python
## 2. Create model object

# Model object
class SoftMax(nn.Module):
    
    # Constructor
    def __init__(self, input_size, output_size):
        super(SoftMax, self).__init__()
        self.linear = nn.Linear(input_size, output_size)
        
    # Prediction
    def forward(self, x):
        z = self.linear(x)
        return z
        
# Model class
model = SoftMax(1,3)

# View initial parameters
print("The parameters: ", list(model.parameters()))
```

```python
## 2. Create a criterion from nn module.
criterion = nn.CrossEntropyLoss()
```

```python
## 3. Create objects for training

# Trainer function
def train_model(criterion, train_loader, optimizer, epochs):    
    LOSS = []
    loss_epoch = 0
    for epoch in range(epochs):
        for x,y in train_loader:
            yhat = model(x)
            loss = criterion(yhat, y)
            loss_epoch = loss_epoch + loss
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        LOSS.append(loss_epoch.item())
        loss_epoch = 0
    return LOSS 

# Criterion object 
criterion = nn.CrossEntropyLoss()

# DataLoader object 
train_loader = DataLoader(dataset = data_set, batch_size = 5)

# Optimizer
optimizer = optim.SGD(model.parameters(), lr = 0.1)
```

```python
## 4. Train model
LOSS = train_model(criterion, train_loader, optimizer, epochs = 100)

# View Loss
plt.plot(LOSS)
plt.xlabel("Epochs")
plt.ylabel("Cost")
```

```python
## Get labels 

z =  model(data_set.x)
_, labels = z.max(1)
labels
```

```python
## Get probabilities of belonging to each class

# Create a softmax object
f_softmax = nn.Softmax(dim = -1)
# Calculate probabilities
probs = f_softmax(z)
# Trun tensor into a numpy array use detach to not require grad
probs = probs.detach().numpy().round(5)
# Print in a readable way
np.set_printoptions(suppress = True)
print(probs)
```

```python
# Print the accuracy
correct = (data_set.y == yhat).sum().item()
accuracy = correct / len(data_set)
print("The accuracy: ", accuracy)
```

### Softmax Regression for image classification


The MNIST database (Modified National Institute of Standards and Technology database) is a large database of handwritten digits that is commonly used for training various image processing systems. The database is also widely used for training and testing in the field of machine learning. The MNIST database contains 60,000 training images and 10,000 testing image of handwritten digits ranging from 0 to 9.

Since each image is a greyscale image, the intensity values for each pixel can range from 0 to 255. Further, each image in the MNIST dataset comprises of 784 pixels,thus a single vector image has 784 values in it.

The first element of the tuple is the actual MNIST image loaded as a tensor, and the second element of the tuple is the actual class of the image. 

The first element of the tuple is a 28 by 28 float tensor of the imageand the second element contains a long tensor which is the actual class of the image. We’ll concatenate our 28 by 28 images to a 1D vector. Our images are 28 by 28 dimensions our input dimension will be 28 times 28 to the SoftMax classifier. Also, since we have 10 output classes our output dimension will be 10.

```python
## -1. Auxiliary functions

# The function to plot parameters
def PlotParameters(model): 
    W = model.state_dict()['linear.weight'].data
    w_min = W.min().item()
    w_max = W.max().item()
    fig, axes = plt.subplots(2, 5)
    fig.subplots_adjust(hspace=0.01, wspace=0.1)
    for i, ax in enumerate(axes.flat):
        if i < 10:
            
            # Set the label for the sub-plot.
            ax.set_xlabel("class: {0}".format(i))

            # Plot the image.
            ax.imshow(W[i, :].view(28, 28), vmin=w_min, vmax=w_max, cmap='seismic')

            ax.set_xticks([])
            ax.set_yticks([])

    plt.show()
```

```python
## 0. Import required tools
import torchvision.datasets as dsets
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from torch import nn, optim
```

```python
## 1. Create Dataset object

# Load training Dataset
train_dataset = dsets.MNIST(root = '/Users/RamsesLg/Desktop/Espacio_de_trabajo/Guía_DS/Python/Data', train=True, download=True, transform = transforms.ToTensor())
# View training Dataset
print(train_dataset)
print(type(train_dataset))

# Load validation dataset
validation_dataset = dsets.MNIST(root='/Users/RamsesLg/Desktop/Espacio_de_trabajo/Guía_DS/Python/Data', train=False, download=True, transform = transforms.ToTensor())
# View validation Dataset
print(validation_dataset)
print(type(validation_dataset))
```

```python
# View label of the fifth element
train_dataset[4][1]
```

```python
# View image of the fifth element
plt.figure(figsize=(2.5,2.5))
plt.imshow(train_dataset[4][0].numpy().reshape(28, 28), cmap='gray')
plt.show()
```

```python
## 2. Create a model object using class linear

# Create model class
class SoftMax(nn.Module):
    
    # Constructor
    def __init__(self, input_size, output_size):
        super(SoftMax, self).__init__()
        self.linear = nn.Linear(input_size, output_size)
        
    # Prediction
    def forward(self, x):
        z = self.linear(x)
        return z
        
# Create model object
input_dim = 28 * 28
output_dim = 10
model = SoftMax(input_dim, output_dim)

# Print the parameters size
print('W: ',list(model.parameters())[0].size())
print('b: ',list(model.parameters())[1].size())
```

```python
# Plot initial parameetrs
PlotParameters(model)
```

```python
## 3. Create objects for training

# Training function
# Train and validate per epoch
def train_model(criterion, train_loader, validation_loader, optimizer, epochs):
    loss_list = []
    accuracy_list = []
    n_test = len(validation_dataset)
    for epoch in range(epochs):
        for x, y in train_loader:
            optimizer.zero_grad()
            z = model(x.view(-1, 28 * 28))
            loss = criterion(z, y)
            loss.backward()
            optimizer.step()
            
        correct = 0
        # perform a prediction on the validationdata  
        for x_test, y_test in validation_loader:
            z = model(x_test.view(-1, 28 * 28))
            _, yhat = torch.max(z.data, 1)
            correct += (yhat == y_test).sum().item()
        accuracy = correct / n_test
        loss_list.append(loss.data)
        accuracy_list.append(accuracy)
    return loss_list, accuracy_list
        
# Criterion function
criterion = nn.CrossEntropyLoss()

# DataLoader objects
train_loader = torch.utils.data.DataLoader(dataset = train_dataset, batch_size = 100)
validation_loader = torch.utils.data.DataLoader(dataset = validation_dataset, batch_size = 5000)

# Optimizer object
optimizer = torch.optim.SGD(model.parameters(), lr = 0.1)
```

```python
## 4. Train model

# Train model
loss_list, accuracy_list = train_model(criterion, train_loader, validation_loader, optimizer, epochs = 100)

# Plot loss and accuracy
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.plot(loss_list,color=color)
ax1.set_xlabel('epoch',color=color)
ax1.set_ylabel('total loss',color=color)
ax1.tick_params(axis='y', color=color)
    
ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('accuracy', color=color)  
ax2.plot(accuracy_list, color=color)
ax2.tick_params(axis='y', color=color)
fig.tight_layout()
```

```python
# Plot final parameters
PlotParameters(model)
```
